```python
# Liberies:
!pip install spectral
from spectral import *
import numpy as np
import pandas as pd
```

```python
def HSI_to_df(path_to_image,list_of_columns):
  img = open_image(path_to_image) # Read the image
  y = img[:,:,:].shape[0] # y axis
  x = img[:,:,:].shape[1] # x axis
  z = img[:,:,:].shape[2] # z axis (bands)
  reshaped= img[:,:,:].reshape(x*y,z)  # Reshape to 2D array
  df = pd.DataFrame(reshaped, columns=list_of_columns) #Convert to df
  
  return df
```
